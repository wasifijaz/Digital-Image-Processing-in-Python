```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('shape.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray, 47, 0.01, 20)
corners = np.int0(corners)
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(img,(x,y),3,255,-1)
```

```python
plt.imshow(img)
plt.show()
```
