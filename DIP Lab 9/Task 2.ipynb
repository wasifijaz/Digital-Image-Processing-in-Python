```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('pic.jpg')
cv2.imshow('image', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(img)
```

```python
imgNew = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
cv2.imshow('image', imgNew)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(imgNew)
```
