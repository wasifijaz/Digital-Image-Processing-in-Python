```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('pic.jpg',0)
window_name = 'img'
cv2.imshow(window_name,image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
histogram = cv2.calcHist([image],[0],None,[256],[0,256])
plt.plot(histogram, color='k')
plt.show()
```

```python
img = cv2.imread('pic.jpg')
for i,col in enumerate(['b','g','r']):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])
    plt.show()
```

```python
gray_img =  cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray_img_eqhist =  cv2.equalizeHist(gray_img)
```

```python
window_name = 'img'
cv2.imshow(window_name,gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.plot(gray_img_eqhist, color='k')
plt.show()
```

```python
ret = cv2.threshold(gray_img,80,256,cv2.THRESH_BINARY)
```

```python
flip = cv2.flip(img,0)
window_name = 'img'
cv2.imshow(window_name,flip)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blurImg = cv2.blur(img,(20,20))
window_name = 'img'
cv2.imshow(window_name,blurImg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gausBlur = cv2.GaussianBlur(img,(5,5),0)
window_name = 'img'
cv2.imshow(window_name,gausBlur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
medBlur = cv2.medianBlur(img,5)
window_name = 'img'
cv2.imshow(window_name,medBlur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
bilFilter = cv2.bilateralFilter(img,9,75,75)
window_name = 'img'
cv2.imshow(window_name,bilFilter)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('pic2.jpg', flags=cv2.IMREAD_COLOR)
window_name = 'imgBlur'
cv2.imshow(window_name,image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
image_sharp = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)
cv2.imshow(window_name, image_sharp)
cv2.waitKey()
cv2.destroyAllWindows()
```
